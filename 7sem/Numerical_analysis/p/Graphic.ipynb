```python
from matplotlib import pylab as plt
from matplotlib.ticker import FuncFormatter
from pylab import rcParams
import numpy as np
from math import ceil


%matplotlib inline

rcParams['figure.figsize'] = 20, 5
y_min = -1
y_max = 3
x_min = 0
x_max = 10

def parting(xs, parts):
    part_len = ceil(len(xs)/parts)
    return [xs[part_len*k:part_len*(k+1)] for k in range(parts)]

delta_t = 0.1
delta_x = 0.1

def x_format(x, pos):
    return '%d' % int(x*delta_x)

def y_format(y, pos):
    return '%d' % int(y*delta_t)

formatter_x = FuncFormatter(x_format)
formatter_y = FuncFormatter(y_format)
```

```python
##Просто график функций##

x = []
p = []
v = []

with open("../data/x") as fx:
    for line in fx:
        for i, dot in enumerate(line.split()):
            x.append(float(dot))

with open("../data/p") as fp:
    for line in fp:
        for i, dot in enumerate(line.split()):
            p.append(float(dot))

with open("../data/v") as fv:
    for line in fv:
        for i, dot in enumerate(line.split()):
            v.append(float(dot))
                
plt.ylim(y_min, y_max)
plt.xlim(x_min, x_max)
plt.plot(x, p)
plt.plot(x, v)
plt.legend ( ("плотность", "скорость") )
plt.show()
```

```python
##Гистограмма плотности##
x = []
z = []

with open("../data/x") as fx:
    for line in fx:
        for i, dot in enumerate(line.split()):
            x.append(float(dot))

with open("../data/p_hist") as hist:
    for line in hist:
        for i, dot in enumerate(line.split()):
            z.append(float(dot))

z_count = len(z)
x_count = len(x)
t_count = int(z_count/x_count)


Z = np.array(parting(z, t_count))


fig, ax = plt.subplots(nrows=1)

im = ax.pcolormesh(Z.transpose())

fig.colorbar(im, ax=ax)

ax.set_xlabel('Время')
ax.set_ylabel('Пространство')
ax.yaxis.set_major_formatter(formatter_y)
ax.xaxis.set_major_formatter(formatter_x)



fig.tight_layout()
plt.show()
```

```python
##Гистограмма скорости##
x = []
z = []

with open("../data/x") as fx:
    for line in fx:
        for i, dot in enumerate(line.split()):
            x.append(float(dot))

with open("../data/v_hist") as hist:
    for line in hist:
        for i, dot in enumerate(line.split()):
            z.append(float(dot))

z_count = len(z)
x_count = len(x)
t_count = int(z_count/x_count)


Z = np.array(parting(z, t_count))


fig, ax = plt.subplots(nrows=1)

im = ax.pcolormesh(Z.transpose())

fig.colorbar(im, ax=ax)

ax.set_xlabel('Время')
ax.set_ylabel('Пространство')
ax.yaxis.set_major_formatter(formatter_y)
ax.xaxis.set_major_formatter(formatter_x)



fig.tight_layout()
plt.show()
```
