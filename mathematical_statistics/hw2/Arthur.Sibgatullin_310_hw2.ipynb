<h3>Геометрическое распределение - $Geom(p)$</h3>

__Определение:__ распределение вероятностей случайной величины $\displaystyle{X}$ равной номеру первого «успеха» в серии испытаний Бернулли и принимающей значения n=1,2,3,... называется геометрическим распределением (обозначается $Geom(p)$). Ф-ия распределния $P_{Geom(p)}(X = x_i) = (1-p)^{x_i-1}p$  

<h4>1.Метод моментов</h4>

$\alpha_1 = E(\xi) = 1 / p. Пусть \bar{x}$ - оценка для $E(\xi)$, тогда $\bar{x} = \frac{1}{\widehat{p}} \, => \widehat{p} = \frac{1}{\bar{x}}$  
__Ответ:__ $\widehat{p} = \frac{1}{\bar{x}}$
<h4>2.Метод максимума правдоподобия</h4>  

Функция правдоподобия:$$L(\bar{X},p) = \prod_{i=1}^n{P(X_i, p)} = p^n\prod_{i=1}^n{(1-p)^{x_i - 1}}$$. Для удобства будем рассматривать $ln(L(\bar{X},p))$, т.к. $ln(x)$ - монотонная функция.  

$$ln(L(\bar{X},p)) = ln [p^n\prod_{i=1}^n{(1-p)^{x_i - 1}}] = n\,ln(p) + \sum_{i=1}^n{ln ((1-p)^{x_i - 1})} = n\,ln(p) - n\,ln(1-p) +ln(1-p)\sum_{i=1}^n{x_i}$$

Найдем при каком $p$ достигается максимум $ln(L(\bar{X},p))$.  

$$\frac{\partial\, ln(L(\bar{X},p)}{\partial \,p} = 0 => \frac{n}{p} + \frac{\sum_{i=1}^n{x_i}}{p - 1} - \frac{n}{p-1} = 0$$.  
Выразим $p$.

$$p = \frac{n}{\sum_{i=1}^n{x_i}} => \widehat{p} = \frac{1}{\bar{x}}$$
Проверим, что при данном $p$ действительно достигается максимум.  
$$\left. \frac{\partial^2\, ln(L(\bar{X},p)}{\partial \,p^2} \right|_{p=\frac{1}{\bar{X}}} = -n (\bar{X})^2 - (\sum_{i=1}^n{x_i}) * (\bar{X} - 1)^2 + n (\bar{X} - 1)^2 < 0$$. 

__Ответ:__ $\widehat{p} = \frac{1}{\bar{x}}$


<h3>Сдвинутое показательное распределение</h3>

__Определение:__ Сдвинутым распределением называется распределние с данной плотностью:
$p(x) = \frac{1}{\theta}e^{-\frac{1}{\theta}(x-x_0)},\,\, x\ge x_0$. 

Необходимо построить оценку для параметра $(x_0, \theta)$ 

<h4>1.Метод моментов</h4>
$$\alpha_1 = E(\xi) = \int_{-\infty}^{+\infty}{xp(x) dx} = \int_{x_0}^{+\infty}{\frac{x}{\theta}\,e^{-\frac{(x-x_0)}{\theta}} dx}$$  
Сделаем замену $u = \frac{x-x_0}{\theta}$, тогда $x = \theta u + x_0$ и $dx = \theta du$, и посчитаем неопределенный интеграл с такой заменой:  
$$\int{\frac{x}{\theta}\,e^{-\frac{(x-x_0)}{\theta}} dx} = \int{(\theta u + x_0)\,e^{-u} du} = x_0\int{e^{-u} du} + \theta \int{u \,e^{-u} du} = -e^{-u} (\theta u - \theta - x_0) + Const = -e^{-\frac{x-x_0}{\theta}}(x + \theta) + Const $$
Подставив пределы интегрирования получим, что $\alpha_1 = x_0 + \theta$

$$\alpha_2 = E(\xi^2) = \int_{-\infty}^{+\infty}{x^2 p(x) dx} = \int_{x_0}^{+\infty}{\frac{x^2}{\theta}\,e^{-\frac{(x-x_0)}{\theta}} dx}$$
Сделаем замену $u = \frac{x-x_0}{\theta}$, тогда $x = \theta u + x_0$ и $dx = \theta du$, и посчитаем неопределенный интеграл с такой заменой:
$$\int{\frac{x^2}{\theta}\,e^{-\frac{(x-x_0)}{\theta}} dx} = \int{(\theta u + x_0)^2\,e^{-u} du} = x_0^2\int{e^{-u} du} + 2\theta x_0 \int{u \,e^{-u} du} + \theta^2 \int{u^2 \,e^{-u} du} = -e^{-u} (x_0^2 + 2\theta x_0(1 + u) + \theta^2 (u^2 + 2u + 2)) + Const = -e^{-\frac{x-x_0}{\theta}}(x + 2\theta x + 2\theta^2) + Const $$
Подставив пределы интегрирования получим, что $\alpha_2 = x_0^2 +2x_0\theta + 2\theta^2$  

Составим систему уравнений:
$$\begin{cases}
E = \alpha_1 \\
D = \alpha_2 - \alpha_1^2 \\
\end{cases} $$

$$\begin{cases}
E = x_0 + \theta \\
D = \theta^2 \\
\end{cases} $$

$$\begin{cases}
x_0 = E - \sqrt{D} \\
\theta = \sqrt{D} \\
\end{cases} $$

$$\begin{cases}
\widehat{x_0} = \bar{X} - \sqrt{S^2} \\
\widehat{\theta} = \sqrt{S^2} \\
\end{cases} $$, где $S^2$ несмещенная оценка для дисперсии, а $\bar{X}$ - оценка для матожидания

<h4>2.Метод максимума правдоподобия</h4>  
Функция правдоподобия:$$L(\bar{X},(x_0, \theta)) = \prod_{i=1}^n{p(X_i, (x_0, \theta))} = \prod_{i=1}^n{\frac{e^{-\frac{x_i - x_0}{\theta}}}{\theta}} = \frac{1}{\theta^n} e^{\frac{nx_0 - \sum_{i=1}^n{X_i}}{\theta}} =\begin{cases}
\frac{1}{\theta^n} e^{\frac{nx_0 - \sum_{i=1}^n{X_i}}{\theta}} & X_{(1)}\ge x_0\\
0, &\mbox{иначе}
\end{cases}$$
Нерегулярный случай: Заметим что максимум достигается при $\widehat{x_0} = \bar{X}$  

Для удобства будем рассматривать $ln(L(\bar{X},p))$, т.к. $ln(x)$ - монотонная функция.

$\ln L=-n\cdot\ln\theta+\frac{n\cdot x_0}{\theta}-\frac{n\cdot \bar{ x}}{\theta}$  
$\frac{\partial \ln L}{\partial \theta}=-\frac{n}{\theta}-\frac{nx_0}{\theta ^2}+\frac{n\bar {x}}{\theta^2}$

$\widehat{\theta}=\bar{x} - \widehat{x_0} = 0$

<h4> 2 - Вероятность успеха биномиального распределения с неизвестным числом испытаний</h4>


Найдем оценки для $p$ и $n$ через метод моментов:  
$ \overline{X} = \widehat{n}\,\widehat{p}\,,$  
$\widehat{\sigma^2} = \widehat{n}\,\widehat{p}\,(1 - \widehat{p}) $  
$ => \widehat{p} = 1 - \frac{\widehat{\sigma^2}}{\bar{X}} $  
$=> \widehat{n} = \frac{\bar{X}}{\widehat{p}} = \frac{(\bar{X})^2}{\bar{X} - \widehat{\sigma^2}} $


$$ n - \bar{X} = \frac{\bar{X}\hat{\sigma^2}}{\bar{X} - \widehat{\sigma^2}}  \;\; \text{и}\;\; \frac1{n} \left( \bar{X} - u_{\frac{1+\gamma}2}\, \sqrt{\bar{X}\,(n - \bar{X})} \right) = \frac{\bar{X}}{n} (1 - u_{\frac{1+\gamma}2}\, \sqrt{\frac{\hat{\sigma^2}}{\bar{X} - \hat{\sigma^2}}} \;) = \frac{\bar{X} - \hat{\sigma^2}}{\bar{X}} (1 - u_{\frac{1+\gamma}2}\, \sqrt{\frac{\hat{\sigma^2}}{\bar{X} - \hat{\sigma^2}}} \;) => $$

$$=>[ \frac{\bar{X} - \hat{\sigma^2}}{\bar{X}} (1 - u_{\frac{1+\gamma}2}\, \sqrt{\frac{\widehat{\sigma^2}}{\bar{X} - \widehat{\sigma^2}}} \;) \,;\, \frac{\bar{X} - \widehat{\sigma^2}}{\bar{X}} (1 + u_{\frac{1+\gamma}2}\, \sqrt{\frac{\widehat{\sigma^2}}{\bar{X} - \widehat{\sigma^2}}} \;)\;] 
$$
    ассимптотически доверительный интервал уровня $\gamma$


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
import statsmodels.stats.api as sms

def asymp_interval (a, alpha = 0.95):
    Mean  = np.mean(a)
    Size = len (a)
    x = np.std(a, ddof=1)/np.sqrt(Size)
    kv = x * stats.norm.ppf((1 + alpha) / 2)
    return Mean - kv, Mean + kv

data = pd.read_csv ("covid_short_data-5.csv")
age_China = np.array ([data['age'][n] for n in range (len (data['age'])) if not np.isnan(data['age'][n]) and data['country'][n] == 'China'])

print("M: %lf" % age_China.mean())
print("D: %lf" % np.var(age_China, ddof = 1))
print("Confidence interval:")
print(stats.t.interval(0.95, len(age_China)-1, loc=np.mean(age_China), scale=stats.sem(age_China)))
print (asymp_interval(age_China))
```
